# 1. Variáveis e Tipos de Dados

**O Conceito:**
Imagine que você está de mudança para uma casa nova. Você tem várias caixas de papelão vazias. Para saber o que tem dentro, você escreve uma etiqueta na caixa com uma caneta, e depois coloca um objeto dentro.

Em Python:
* **A Variável** é a Caixa.
* **O Valor** é o que está dentro da Caixa.
* **O Nome** é a etiqueta da Caixa.

**Regras de Nomes:**
Nomes de variáveis não podem ter espaços! Usamos `snake_case` (palavras separadas por underline/sublinhado).
* *Ruim:* `meu ponto = 10`
* *Bom:* `meu_ponto = 10`

**Os Três Grandes Tipos de Dados:**
O computador precisa saber que tipo de dado está na caixa.
1.  **Inteiro (`int`):** Números inteiros. (ex: `5`, `-10`, `42`). SEM casas decimais.
2.  **Float / Ponto Flutuante (`float`):** Números "quebrados". (ex: `3.14`, `9.99`, `0.001`). TEM casas decimais.
3.  **String / Texto (`str`):** Texto. Deve estar sempre cercado por aspas `" "` ou `' '`.

* **Documentação Oficial:** [Variáveis em Python](https://docs.python.org/pt-br/3/tutorial/introduction.html#using-python-as-a-calculator)

# 3. Input: Ouvindo o Usuário

**O Conceito:**
Até agora, nossos programas são solitários — eles só falam, não escutam. O comando `input()` permite que o usuário (você) digite algo para o programa.

**A Pergunta (Prompt):**
Você deve sempre colocar um texto (string) dentro dos parênteses para o usuário saber o que digitar: `input("Digite aqui: ")`.

**AVISO CRÍTICO - A ARMADILHA DO TEXTO:**
Quando você usa `input()`, o computador *sempre* trata a resposta como texto (String), mesmo que o usuário digite um número.
* Se o usuário digita `5`, o computador vê `"5"` (texto).
* `"5" + "5"` resulta em `"55"` (textos colados), e não `10` (matemática).

**A Solução:**
Você deve **converter** a entrada se quiser usá-la como matemática.
* `int( input(...) )` -> Transforma em número inteiro.
* `float( input(...) )` -> Transforma em número decimal.

* **Documentação Oficial:** [A Função Input](https://docs.python.org/pt-br/3/library/functions.html#input)

````python
# EXEMPLO 1: Entrada de Texto Simples
# Nenhuma conversão necessária aqui, pois nomes são texto.
nome = input("Qual é o seu nome? ")
print("Prazer em te conhecer,", nome)
````

````python
# EXEMPLO 2: Entrada de Inteiro (O Jeito Certo)
# Nós "embrulhamos" o input com int()
idade = int(input("Quantos anos você tem? "))
print("Ano que vem você terá:", idade + 1)
````

````python
# EXEMPLO 3: Entrada de Float (Decimais)
# Preços geralmente têm decimais, então usamos float()
preco = float(input("Quanto custa a pizza? "))
print("O preço é:", preco)
````

**O Conceito:**

**Os Operadores:**
* `+` (Adição/Soma)

````python
# EXEMPLO 3: Expoentes e Ordem
# 2 elevado a 3 é 8
print(2 ** 3)

# Parênteses importam:
print(2 + 5 * 10)   # 52 (Multiplicação acontece primeiro)
print((2 + 5) * 10) # 70 (Parênteses acontecem primeiro)
````

# 5. Lógica Booleana: Verdadeiro ou Falso?

**O Conceito:**
Computadores não "acham" coisas. Eles sabem se algo é `True` (Verdadeiro) ou `False` (Falso). Este é um tipo de dado especial chamado **Booleano**.

**Operadores de Comparação:**
Usamos estes símbolos para fazer perguntas ao computador:
* `==` (Igual a) -> **ATENÇÃO:** Dois iguais perguntam "é igual?". Um igual (`=`) atribui valor.
* `!=` (Diferente de / Não igual)
* `>` (Maior que)
* `<` (Menor que)
* `>=` (Maior ou igual a)
* `<=` (Menor ou igual a)

* **Documentação Oficial:** [Operações Booleanas](https://docs.python.org/pt-br/3/library/stdtypes.html#boolean-operations-and-or-not)

````python
# EXEMPLO 1: Comparações Básicas
# O computador vai responder True ou False
print(10 > 5)   # 10 é maior que 5? (True)
print(10 < 5)   # 10 é menor que 5? (False)
````

````python
# EXEMPLO 2: O Grande Erro do "Igual"
# Lembre-se: == compara, = define.
meu_numero = 50
numero_vencedor = 50

# Estamos perguntando: "Eles são iguais?"
print(meu_numero == numero_vencedor)
````

````python
# EXEMPLO 3: Textos Diferentes
# O Python diferencia maiúsculas de minúsculas!
print("Python" == "python") # False, porque 'P' é diferente de 'p'
print("A" != "B")           # True, A é diferente de B
````

# 6. Condicionais: Ensinando o Computador a Tomar Decisões

**O Professor Explica:**
Até agora, nosso código corria como um trem em um trilho reto: ele executava a linha 1, depois a 2, depois a 3, sem desvios.
Mas a vida real não é assim. A vida é feita de escolhas.
* "SE estiver chovendo, levo um guarda-chuva. SENÃO, vou de óculos escuros."
* "SE eu tiver a chave, abro a porta. SENÃO, toco a campainha."

Na programação, chamamos isso de **Estrutura Condicional**. Nós usamos as palavras-chave em inglês: `if` (se), `elif` (abreviação de "else if" - senão se) e `else` (senão).



**A Regra de Ouro: Indentação (O Espaço Mágico)**
O Python é muito organizado. Para o computador saber quais linhas de código pertencem ao "SE", nós precisamos empurrar o código para a direita. Chamamos isso de **Indentação**.
* Você deve usar a tecla `Tab` (ou 4 espaços) para criar esse recuo.
* Se o código estiver encostado na margem esquerda, ele está "fora" da condição.
* **Atenção:** Esquecer de indentar é o erro número 1 de iniciantes (Erro: `IndentationError`).

**Como o computador pensa:**
1.  Ele chega no `if`. A condição é verdadeira (`True`)?
2.  **Sim:** Ele entra no bloco indentado, executa o código e **pula** todo o resto da estrutura (ignora os `elif` e `else`).
3.  **Não:** Ele pula o bloco do `if` e testa o `elif`.
4.  Se nada der certo, ele executa o `else` como última opção de segurança.

**Estrutura do Código:**
```python
if tenho_dinheiro:
    print("Vou comprar lanche!")  # Só acontece se for verdade
else:
    print("Vou ficar com fome.")  # Só acontece se for mentira
print("Vida que segue.")          # Acontece sempre (não está indentado)
```

* **Documentação Oficial:** [Ferramentas de Controle de Fluxo](https://docs.python.org/pt-br/3/tutorial/controlflow.html#if-statements)

````python
# EXEMPLO 1: O 'If' Simples
bateria = 10

if bateria < 20:
    print("Aviso: Bateria fraca!") # Esta linha só roda se a condição for True
````

````python
# EXEMPLO 2: If e Else (Sim ou Não)
senha = input("Digite a senha: ")

if senha == "1234":
    print("Acesso Permitido.")
else:
    print("Acesso Negado.")
````

````python
# EXEMPLO 3: If, Elif e Else (Múltiplas opções)
# O computador testa na ordem. O primeiro que for verdadeiro, ganha.
nota = 85

if nota >= 90:
    print("Nota A - Excelente!")
elif nota >= 80:
    print("Nota B - Muito bom!")
else:
    print("Precisa estudar mais.")
````

# 7. Loops: O Poder da Repetição Infinita

**O Professor Explica:**
Imagine que você cometeu um erro na escola e o professor pediu para você escrever no quadro negro 100 vezes: *"Eu prestarei atenção na aula de Python"*.
Se você fosse humano, isso levaria horas e sua mão doeria.
Para um computador, isso leva 0.001 segundos.

Na programação, quando queremos repetir uma ação várias vezes, não copiamos e colamos o código (isso é proibido!). Nós usamos um **Loop** (Laço de Repetição).
O loop mais famoso e útil em Python é o **`for`** (para).



# Visualizando o Loop For

Muitas vezes, uma imagem vale mais que mil palavras.
Aqui está como o computador "pensa" quando vê um loop `for`:

      +----------------+

      +----------------+

    |  Pegar 1º Item da  | <----.

      +----------------+


**A Mágica do `range()`:**
Para dizer ao computador "quantas vezes" repetir, usamos a função `range()`. Ela cria uma sequência de números para o loop percorrer.
A função `range` tem 3 segredos:
1.  **O Fim é Exclusivo:** Se você escrever `range(5)`, o computador conta: 0, 1, 2, 3, 4. Ele **PARA** antes de chegar no 5. O 5 nunca é incluído.
2.  **Começa no Zero:** Se você não disser onde começar, ele sempre começa do 0.
3.  **A Variável 'i':** No código `for i in range(5):`, a letra `i` é uma variável especial (o contador). A cada volta que o loop dá, o `i` muda de valor automaticamente.

**Estrutura do Código:**
```python
for i in range(5):
    print("Isso se repete.")
    print("O valor de i agora é:", i)
```

````python
# EXEMPLO 1: A Repetição Básica
# Vamos imprimir uma frase 5 vezes.
# Note que, assim como no IF, precisamos INDENTAR (dar espaço) no código dentro do loop.

print("Iniciando a repetição...")

for i in range(5):
    print("Eu amo programar em Python!")

print("Fim da repetição.")
````

````python
# EXEMPLO 2: O Poder do Contador 'i'
# Vamos ver como a variável 'i' muda sozinha a cada volta.

for numero in range(0, 4):
    print("Estamos na volta número:", numero)
    print("---")
````

````python
# EXEMPLO 3: O Range Personalizado (Início e Fim)
# Queremos contar os capítulos de um livro, de 1 a 5.
# Lembre-se: Para chegar no 5, o range precisa ir até 6!

print("Índice do Livro:")

for capitulo in range(1, 6):
    print("Capítulo", capitulo)
````

# 8. O Operador Módulo (%): O Segredo do Resto

**O Professor Explica:**
Muitos alunos olham para o símbolo `%` e pensam: *"Ah, porcentagem!"*.
**PARE!** Em Python, o `%` **NÃO** é porcentagem.
Ele é o Operador de **Módulo** (ou Resto da Divisão).



[Image of long division showing quotient and remainder]


**A Matemática da 4ª Série:**
Lembra quando você aprendeu a dividir antes de usar vírgulas?
* Imagine que você tem **10 fatias de pizza** e **3 amigos**.
* Você dá 3 fatias para cada um. (3 x 3 = 9).
* Sobra **1 fatia** na caixa.
* Em Python, dizemos: `10 % 3` é igual a `1`.

**A Regra de Ouro da Programação:**
Para que serve isso? Para descobrir padrões!
Se o resto da divisão de um número `A` por `B` for **ZERO** (`0`), significa que a divisão foi perfeita. Ou seja, `A` é divisível por `B`.

**Aplicações Práticas:**
1.  **Par ou Ímpar:** Se um número dividido por 2 deixa resto 0, ele é PAR. Se deixa resto 1, é ÍMPAR.
2.  **Múltiplos:** Para saber se um número é múltiplo de 5, fazemos `numero % 5 == 0`.
3.  **Tempo:** Para calcular segundos em minutos (65 segundos = 1 minuto e sobra 5 segundos).

* **Documentação Oficial:** [Matemática em Python](https://docs.python.org/pt-br/3/reference/expressions.html#binary-arithmetic-operations)

````python
# EXEMPLO 1: Vendo o Resto na Prática
# Vamos ver o que sobra dessas divisões.

print("10 dividido por 2. Resto:", 10 % 2)  # 0 (Divisão exata)
print("10 dividido por 3. Resto:", 10 % 3)  # 1 (Sobra 1)
print("10 dividido por 4. Resto:", 10 % 4)  # 2 (Sobram 2)
print("10 dividido por 5. Resto:", 10 % 5)  # 0 (Divisão exata)
````

````python
# EXEMPLO 2: O Verificador de Par ou Ímpar
# Esta é a lógica mais usada no mundo.

numero = 13

if numero % 2 == 0:
    print(numero, "é PAR.")
else:
    print(numero, "é ÍMPAR (porque o resto é 1).")
````

````python
# EXEMPLO 3: Encontrando Múltiplos
# Vamos verificar se um número é "especial" (múltiplo de 10).

pontos = 50

if pontos % 10 == 0:
    print("Parabéns! Você atingiu uma pontuação redonda!")
else:
    print("Sua pontuação é quebrada.")
````

# 9. O Desafio Final: O Teste FizzBuzz

**O Professor Explica:**
Meus parabéns por chegarem até aqui.
Agora, você vai enfrentar um teste real. O "FizzBuzz" é um problema famoso usado em entrevistas de emprego no Google, Amazon e Microsoft para filtrar candidatos. Muitos programadores falham nisso porque não prestam atenção na lógica.

**A Missão:**
Escreva um código que conte de **1 até 100**.

**As Regras do Jogo:**
Para cada número nessa contagem, o computador deve decidir o que imprimir:
1.  Se o número for divisível por **3 e 5** (ao mesmo tempo), imprima **"FizzBuzz"**.
2.  Se o número for divisível apenas por **3**, imprima **"Fizz"**.
3.  Se o número for divisível apenas por **5**, imprima **"Buzz"**.
4.  Se não for divisível por nenhum deles, imprima apenas **o próprio número**.

**Exemplo de como deve ficar a saída (o console):**
```text
1
2
Fizz
4
Buzz
Fizz
7
...
14
FizzBuzz
```

**Dicas Estratégicas do Professor (Leia com Atenção!):**

* **Ferramentas:** Você vai precisar de um Loop `for` com `range`. Lembre-se que o `range(1, 100)` para no 99. Como fazer ele ir até 100?
* **A Ordem Importa:** O computador lê o código de cima para baixo.
    * Imagine o número **15**. Ele é divisível por 3? Sim. Ele é divisível por 5? Sim.
    * Se você perguntar primeiro *"É divisível por 3?"*, o computador vai dizer SIM, imprimir "Fizz" e pular o resto. Ele nunca vai chegar no "FizzBuzz".
* **Dica de Ouro:** Teste a condição mais difícil/específica primeiro (o `and`).

Boa sorte, programadores.

````python
# --- TAREFA FINAL: FIZZBUZZ ---
# Nome do Aluno: Maria Beatriz

# DICA: Comece criando o Loop.
# Depois, coloque a estrutura de If / Elif / Else dentro dele.

# Escreva seu código abaixo desta linha:


for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````
