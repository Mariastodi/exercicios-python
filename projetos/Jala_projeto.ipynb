```python
class Livro:
    def __init__(self, titulo, autor, genero, ano):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.ano = ano

    def __str__(self):
        return f'Título: {self.titulo} Autor: {self.autor} Gênero: {self.genero} Ano: {self.ano}'


class Biblioteca:
    def __init__(self):
        self.livros = []

    def adicionar_livro(self, livro):
        self.livros.append(livro)

    def adicionar_livro_interativamente(self):
        print("Adicionar novo livro à biblioteca:")
        titulo = input("Título: ")
        autor = input("Autor: ")
        genero = input("Gênero: ")
        try:
            ano = int(input("Ano de publicação: "))
        except ValueError:
            print("Ano inválido. Tente novamente.")
            return

        novo_livro = Livro(titulo, autor, genero, ano)
        self.adicionar_livro(novo_livro)
        print(f"Livro '{titulo}' adicionado com sucesso!")

    def listar_livros(self):
        if not self.livros:
            print("Nenhum livro disponível.")
        else:
            print("Livros Disponíveis:")
            for livro in self.livros:
                print(f"- {livro}")

    def buscar_livro_por_titulo(self, titulo):
        for livro in self.livros:
            if livro.titulo.lower() == titulo.lower():
                return livro
        return None

    def remover_livro(self, livro):
        if livro in self.livros:
            self.livros.remove(livro)


class Leitor:
    def __init__(self, nome, id_leitor):
        self.nome = nome
        self.id_leitor = id_leitor
        self.livros_lidos = []

    def ler_livro(self, biblioteca):
        titulo = input('Digite o título do livro que deseja ler: ')
        livro = biblioteca.buscar_livro_por_titulo(titulo)
        if livro:
            print(f'{self.nome} está lendo: {livro.titulo}.')
            self.livros_lidos.append(livro)
            biblioteca.remover_livro(livro)
        else:
            print("Livro não encontrado ou não disponível.")

    def devolver_livro(self, biblioteca):
        if not self.livros_lidos:
            print("Você não tem livros para devolver.")
            return

        print("Livros com você:")
        for i, livro in enumerate(self.livros_lidos):
            print(f"{i + 1} - {livro}")

        try:
            escolha = int(input("Digite o número do livro a ser devolvido: ")) - 1
            if 0 <= escolha < len(self.livros_lidos):
                livro = self.livros_lidos.pop(escolha)
                biblioteca.adicionar_livro(livro)
                print(f"{self.nome} devolveu: {livro.titulo}")
            else:
                print("Opção inválida.")
        except ValueError:
            print("Entrada inválida.")

    def menu_inicial(self, biblioteca):
        while True:
            print(f'\nOlá, {self.nome}.')
            print("1 - Ver Livros Disponíveis")
            print("2 - Pegar Livro")
            print("3 - Devolver Livro")
            print("4 - Adicionar Livro à Biblioteca")
            print("5 - Sair")
            opcao = input("Escolha uma opção: ")

            if opcao == '1':
                biblioteca.listar_livros()
            elif opcao == '2':
                self.ler_livro(biblioteca)
            elif opcao == '3':
                self.devolver_livro(biblioteca)
            elif opcao == '4':
                biblioteca.adicionar_livro_interativamente()
            elif opcao == '5':
                print("Até logo!")
                break
            else:
                print("Opção inválida. Tente novamente.")


def iniciar_sistema():
    biblioteca = Biblioteca()

    biblioteca.adicionar_livro(Livro("1984", "George Orwell", "Distopia", 1949))
    biblioteca.adicionar_livro(Livro("Dom Casmurro", "Machado de Assis", "Romance", 1899))
    biblioteca.adicionar_livro(Livro("O Hobbit", "J.R.R. Tolkien", "Fantasia", 1937))

    print("Bem-vindo à Biblioteca Virtual")
    nome = input("Digite seu nome: ")
    id_leitor = input("Digite seu ID de leitor: ")

    leitor = Leitor(nome, id_leitor)
    leitor.menu_inicial(biblioteca)


iniciar_sistema()
```
