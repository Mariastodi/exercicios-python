Seu objetivo é desenhar formas geométricas usando apenas texto (asteriscos `*` e espaços ` `). Embora pareça simples, isso exige que você entenda como controlar o fluxo do seu código linha a linha.

## Conceito Importante: A função `range()`

Para resolver estes exercícios, você precisará dominar a função `range()`. Ela é usada para gerar sequências de números, essenciais para controlar nossos loops `for`.

A estrutura básica é: `range(inicio, fim, passo)`

1.  **Básico (Apenas o fim):**
    * `range(5)` → Gera: `0, 1, 2, 3, 4`.
    * *Nota:* O número final **não** é incluído. Começa no 0 por padrão.

2.  **Início e Fim:**
    * `range(1, 6)` → Gera: `1, 2, 3, 4, 5`.
    * Útil para quando você quer começar a contar de 1 asterisco, e não de 0.

3.  **Passo (Pular números):**
    * `range(1, 10, 2)` → Gera: `1, 3, 5, 7, 9`.
    * Isso é muito útil para pirâmides centralizadas, que usam números ímpares de asteriscos (*, ***, *****).

4.  **Decrescente (Contagem regressiva):**
    * `range(5, 0, -1)` → Gera: `5, 4, 3, 2, 1`.
    * Aqui, o "passo" é negativo. Fundamental para pirâmides invertidas!

**Execute as células abaixo para ver esses exemplos na prática:**

````python
print("--- Exemplo 1: Simples ---")
for i in range(5):
    print(i, end=" ") # o end=" " imprime na mesma linha
````

````python
print("--- Exemplo 2: Ímpares (Passo 2) ---")
for i in range(1, 10, 2):
    print(i, end=" ")
````

````python
print("--- Exemplo 3: Contagem Regressiva (Passo -1) ---")
for i in range(5, 0, -1):
    print(i, end=" ")
````

### 1. Pirâmide Crescente (Alinhada à Esquerda)
Peça ao usuário um número inteiro (que será a altura da pirâmide) e construa o formato abaixo.

**Exemplo (se o usuário digitar 6):**

```
*
**
***
****
*****
******
```

````python
%reset -f
# Escreva aqui a sua solução

altura = int(input("Qual a altura da sua pirâmide? "))

for linha in range(1, altura + 1):
    for coluna in range(linha):
        print("*", end="")
    print()
````

### 2. Pirâmide Decrescente (Alinhada à Esquerda)
Agora, inverta a lógica. Comece com o número total de asteriscos e vá diminuindo até chegar em 1.

**Exemplo:**
```
******
*****
****
***
**
*
```

````python
%reset -f
# Escreva aqui a sua solução

altura = int(input("Qual a altura da sua pirâmide? "))

for i in range(altura):
    print("*" * (altura - i))
````

### 3. Pirâmide Decrescente (Alinhada à Direita)
Este é um pouco mais complexo. Note que os asteriscos estão "empurrados" para a direita. Isso significa que existem **espaços em branco** invisíveis antes dos asteriscos.

**Exemplo:**

```
******
 *****
  ****
   ***
    **
     *
```

````python
%reset -f
# Escreva aqui a sua solução

altura = int(input("Qual a altura da sua pirâmide? "))

for i in range(altura):
    espacos = i
    estrelas = altura - i
    print(" " * espacos + "*" * estrelas)
````

### 5. A Árvore (Pirâmide Centralizada)
Construa uma pirâmide que cresce para os dois lados, como uma árvore de natal.

**Exemplo:**
```
     *
    ***
   *****
  *******
 *********
***********
```

````python
%reset -f
# Escreva aqui a sua solução

altura = int(input("Qual a altura da sua árvore? "))

for i in range(1, altura + 1):
    espacos = " " * (altura - i)
    estrelas = "*" * (2 * i - 1)
    print(espacos + estrelas)
````

### 5. Pirâmide Invertida (Centralizada)
Aqui a pirâmide diminui dos dois lados, formando um triângulo invertido perfeito.

**Exemplo:**
```
***********
 *********
  *******
   *****
    ***
     *
```

````python
%reset -f
# Escreva aqui a sua solução

altura = int(input("Qual a altura da sua pirâmide invertida? "))

for i in range(altura, 0, -1):
    espacos = " " * (altura - i)
    estrelas = "*" * (2 * i - 1)
    print(espacos + estrelas)
````
