# Desafio 1: O Professor Rigoroso

**O Problema:**
Crie uma calculadora que peça **3 notas** de um aluno (podem ser números quebrados).
1.  Calcule a média das 3 notas.
2.  Se a média for **maior ou igual a 7**, imprima "Aprovado".
3.  Se a média for **menor que 7**, imprima "Reprovado".

**Dica:**
* Use `float(input())` para as notas.
* Lembre-se dos parênteses na média! `(n1 + n2 + n3) / 3`.
* Use `if` e `else`.

````python
%reset -f
nota1 = float(input("Me diga a primeira nota: "))
nota2 = float(input("Me diga a segunda nota: "))
nota3 = float(input("Me diga a terceira nota: "))

media = (nota1 + nota2 + nota3)/3

if media >= 7:
  print("Aprovado")
else:
  print("Reprovado")
````

# Desafio 3: Calculadora de Fatorial

**O Problema:**
Na matemática, o fatorial de um número (representado por `!`) é a multiplicação dele por todos os seus antecessores até 1.
* Exemplo: `5! = 5 * 4 * 3 * 2 * 1 = 120`.

Crie um programa que:
1.  Peça um número ao usuário (ex: 5).
2.  Calcule o fatorial desse número usando um Loop.
3.  Imprima o resultado final.

**Dica de Lógica:**
* Você precisa de uma variável para guardar o resultado. Comece ela com **1** (não 0, senão a multiplicação zera tudo!).
* Use um loop `range(1, numero + 1)`.
* Dentro do loop, atualize a variável: `resultado = resultado * i`.

````python
%reset -f
# Escreva o código do Fatorial aqui:

n = int(input("fala um número ai: "))
resultado = 1

for i in range(1, n + 1):
    resultado = resultado * i

print(resultado)
````

# Desafio 4: É Primo ou não?

**O Problema:**
Um número primo é aquele que só pode ser dividido por 1 e por ele mesmo (ex: 7, 11, 13).
Se ele for divisível por qualquer outro número no meio do caminho, ele NÃO é primo.

Crie um programa que peça um número (ex: 13) e diga se ele é Primo ou Não.

**Dica de Lógica:**
1.  Peça o número ao usuário.
2.  Crie uma variável `divisores = 0`.
3.  Faça um loop de 1 até o número do usuário.
4.  Dentro do loop, se o número do usuário for divisível pelo contador `i` (`numero % i == 0`), aumente a variável `divisores` em +1.
5.  No final (fora do loop), faça um `if`: Se `divisores` for exatamente 2 (só dividiu por 1 e ele mesmo), é primo. Caso contrário, não é.

````python
%reset -f
# Escreva o código de Números Primos aqui:

n = int(input("me diga um numero ai: "))

divisores = 0

for i in range(1, n + 1):
    if n % i == 0:
        divisores += 1
if divisores == 2:
    print("ele é primo")
else:
    print("ele não é primo")
````

# Desafio 5: A Urna Eletrônica

**O Problema:**
Simule uma eleição com 5 eleitores.
Existem dois candidatos: Candidato **10** e Candidato **20**.

1.  Crie um loop que rode 5 vezes (pelos 5 eleitores).
2.  Dentro do loop, peça o voto: "Digite 10 ou 20: ".
3.  Se votar 10, some um ponto para o candidato 1.
4.  Se votar 20, some um ponto para o candidato 2.
5.  No final (fora do loop), verifique quem teve mais votos e imprima: "Vencedor: Candidato X".

````python
%reset -f
# Escreva o código da Urna aqui:
candidato1 = 0
candidato2 = 0

for i in range(5):
  n = int(input("fala ai 10 ou 20: "))
  if n == 10:
    candidato1 += 1

  else:
    candidato2 += 1
if candidato1 > candidato2:
    print("Vencedor: Candidato 10")
elif candidato2 > candidato1:
    print("Vencedor: Candidato 20")
else:
    print("Empate!")
````

# Desafio 6: Números Perfeitos

**O Problema:**
Um número é considerado **Perfeito** se a soma de todos os seus divisores (excluindo ele mesmo) for igual ao próprio número.

* Exemplo: **6** é perfeito.
    * Divisores de 6: 1, 2, 3.
    * Soma: 1 + 2 + 3 = **6**.
* Exemplo: **28** é perfeito.
    * Divisores de 28: 1, 2, 4, 7, 14.
    * Soma: 1 + 2 + 4 + 7 + 14 = **28**.

Seu programa deve pedir um número e dizer: "É Perfeito" ou "Não é Perfeito".

````python
%reset -f
# Escreva o código de Números Perfeitos aqui:

n = int(input("fala um número para testar se o bicho é perfeito: "))
soma = 0

for i in range(1, n):
    if n % i == 0:
        soma += i

if soma == n:
    print("bicho é perfeito")
else:
    print("bicho não é perfeito")
````

# Desafio 7: O Maior e o Menor

**O Problema:**
Peça ao usuário para digitar **5 números** (um por um, dentro de um loop).
No final, o programa deve dizer:
* "O maior número digitado foi: X"
* "O menor número digitado foi: Y"

**Regra:** Você **NÃO** pode usar as funções `max()` ou `min()` do Python. Você deve criar a lógica na mão.

````python
%reset -f
# Escreva o código de Maior e Menor aqui:

n1 = int(input("Digite o primeiro número: "))
maior = n1
menor = n1

for i in range(4):
    num = int(input("Digite o próximo número: "))

    if num > maior:
        maior = num

    if num < menor:
        menor = num

print(f"O maior é {maior}")
print(f"O menor é {menor}")
````
